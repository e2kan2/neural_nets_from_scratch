In this notebook, we will see some activation functions and the derivatives. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(X):
    return 1 / (1 + np.exp(-X))

def sigmoid_grad(X):
    s = sigmoid(X)
    return s * (1 - s)
```

```python
X = np.linspace(-5, 5, 100)

plt.plot(X, sigmoid(X), label="$\sigma(X)$");
plt.plot(X, sigmoid_grad(X), label="$\sigma^\prime(X)$");
plt.grid(True)
plt.legend();
```

```python
X = -10
print(sigmoid(X), sigmoid_grad(X))
X = 0
print(sigmoid(X), sigmoid_grad(X))
X = 10
print(sigmoid(X), sigmoid_grad(X))
```

```python
def tanh(X):
    return np.tanh(X)

def tanh_grad(X):
    t = tanh(X)
    return 1 - t ** 2
```

```python
X = np.linspace(-5, 5, 100)

plt.plot(X, tanh(X), label="$tanh(X)$");
plt.plot(X, tanh_grad(X), label="$tanh^\prime(X)$");
plt.grid(True)
plt.legend();
```

```python
def relu(X):
    return np.maximum(0, X)

def relu_grad(X):
    return np.where(X > 0, 1, 0)
```

```python
X = np.linspace(-5, 5, 100)

plt.plot(X, relu(X), label="$relu(X)$");
plt.plot(X, relu_grad(X), label="$relu^\prime(X)$");
plt.grid(True)
plt.legend();
```

```python
# LReLU'da a!=0.01 ise parametric relu deniyor.

def leaky_relu(X, a=0.1):
    return np.where(X > 0, X, a*X)

def leaky_relu_grad(X, a=0.1):
    return np.where(X > 0, 1, a)
```

```python
X = np.linspace(-5, 5, 100)

plt.plot(X, leaky_relu(X), label="$lrelu(X)$");
plt.plot(X, leaky_relu_grad(X), label="$lrelu^\prime(X)$");
plt.grid(True)
plt.legend();
```

```python
def elu(X, a=0.9):
    return np.where(X > 0, X, a*(np.exp(X) - 1))

def elu_grad(X, a=0.9):
    e = elu(X)
    return np.where(X > 0, 1, e + a)
```

```python
X = np.linspace(-5, 5, 100)

plt.plot(X, elu(X), label="$elu(X)$");
plt.plot(X, leaky_relu_grad(X), label="$elu^\prime(X)$");
plt.grid(True)
plt.legend();
```
